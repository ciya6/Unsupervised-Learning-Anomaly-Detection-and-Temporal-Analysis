**Question 1 : What is Dimensionality Reduction? Why is it important in machine learning?**

**Answer:** Dimensionality reduction is the process of reducing the number of random variables under consideration by obtaining a set of principal variables. It is important in machine learning to combat the 'curse of dimensionality', reduce computational cost, remove redundant features, and improve model performance by simplifying the data representation and reducing overfitting.

**Question 2: Name and briefly describe three common dimensionality reduction techniques.**

**Answer:**
1.  **Principal Component Analysis (PCA):** A linear dimensionality reduction technique that transforms data into a new coordinate system where the greatest variance by any projection comes to lie on the first coordinate (called the first principal component), the second greatest variance on the second coordinate, and so on.
2.  **t-Distributed Stochastic Neighbor Embedding (t-SNE):** A non-linear dimensionality reduction technique used for visualizing high-dimensional data. It models each high-dimensional object by a two- or three-dimensional point in such a way that similar objects are modeled by nearby points and dissimilar objects are modeled by distant points with high probability.
3.  **Linear Discriminant Analysis (LDA):** A supervised dimensionality reduction technique used to find a linear combination of features that characterizes or separates two or more classes of objects or events. It aims to maximize the separability between known categories.

**Question 3: What is clustering in unsupervised learning? Mention three popular clustering algorithms.**

**Answer:** Clustering in unsupervised learning is the task of grouping a set of objects in such a way that objects in the same group (called a cluster) are more similar to each other than to those in other groups. It is an unsupervised task because it does not rely on pre-labeled data.

Three popular clustering algorithms:
1.  **K-Means:** An iterative algorithm that partitions `n` observations into `k` clusters, where each observation belongs to the cluster with the nearest mean (centroid), serving as a prototype of the cluster.
2.  **Hierarchical Clustering:** Builds a hierarchy of clusters. It can be agglomerative (bottom-up, where each observation starts in its own cluster, and pairs of clusters are merged) or divisive (top-down, where all observations start in one cluster, and splits are performed recursively).
3.  **DBSCAN (Density-Based Spatial Clustering of Applications with Noise):** A density-based clustering algorithm that groups together points that are closely packed together, marking as outliers points that lie alone in low-density regions.

**Question 4: Explain the concept of anomaly detection and its significance.**

**Answer:** Anomaly detection (also known as outlier detection) is the process of identifying data points, events, or observations that deviate significantly from the majority of the data. These are often indicators of some underlying problem or rare event. Its significance lies in its ability to detect critical issues such as fraud, system malfunctions, cybersecurity breaches, medical conditions, and unusual patterns in various domains, enabling timely intervention and preventing severe consequences.

**Question 5: List and briefly describe three types of anomaly detection techniques.**

**Answer:**
1.  **Statistical Methods:** These methods assume that normal data instances occur in a high-probability region of a stochastic model, while anomalies occur in a low-probability region. Examples include Gaussian models, Z-score, and Grubbs' test.
2.  **Machine Learning-Based Methods:** These techniques use various machine learning algorithms to learn the patterns of normal data and then identify deviations. Examples include Isolation Forest, One-Class SVM, and Local Outlier Factor (LOF).
3.  **Proximity-Based Methods:** These methods detect anomalies by measuring the distance or density of data points. Anomalies are data points that are far from their nearest neighbors or lie in sparse regions. Examples include k-Nearest Neighbors (k-NN) based anomaly detection.

**Question 6: What is time series analysis? Mention two key components of time series data.**

**Answer:** Time series analysis is a statistical technique used to analyze time series data (data points collected over time) to extract meaningful statistics and other characteristics of the data. It aims to understand the underlying causes of observed patterns, forecast future values, or detect anomalies.

Two key components of time series data are:
1.  **Trend:** The long-term increase or decrease in the data over time.
2.  **Seasonality:** A repetitive pattern or cycle that occurs over a fixed period (e.g., daily, weekly, monthly, yearly).

**Question 7: Describe the difference between seasonality and cyclic behavior in time series.**

**Answer:**
*   **Seasonality:** Refers to patterns in time series data that repeat over a fixed and known period, such as daily, weekly, monthly, or yearly. These patterns are typically caused by factors like weather, holidays, or specific calendar events.
*   **Cyclic Behavior:** Refers to patterns that repeat over periods that are not fixed. The duration of a cycle can vary, and it is usually longer than a seasonal pattern. Cycles are often associated with economic expansions and contractions, or other long-term phenomena that do not have a fixed period.

**Question 8: Write Python code to perform K-means clustering on a sample dataset. (Include your Python code and output in the code box below.)**

**Answer:**

````python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Generate sample data
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Perform K-means clustering
k_means = KMeans(n_clusters=4, init='k-means++', max_iter=300, random_state=0, n_init=10)
clusters = k_means.fit_predict(X)

# Visualize the clusters
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis', s=50, alpha=0.7)
plt.scatter(k_means.cluster_centers_[:, 0], k_means.cluster_centers_[:, 1], s=200, marker='X', c='red', label='Centroids')
plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()

print("Cluster centers:\n", k_means.cluster_centers_)
print("Cluster labels for the first 10 samples:", clusters[:10])
````

**Question 9: What is inheritance in OOP? Provide a simple example in Python.**

**Answer:** Inheritance is a fundamental concept in Object-Oriented Programming (OOP) that allows a class (called a subclass or child class) to inherit attributes and methods from another class (called a superclass or parent class). This promotes code reusability, establishes a natural 'is-a' relationship between classes, and facilitates polymorphism.


```

````python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        raise NotImplementedError("Subclass must implement abstract method")

class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name) # Call the constructor of the parent class
        self.breed = breed

    def speak(self):
        return f"{self.name} says Woof!"

class Cat(Animal):
    def __init__(self, name, color):
        super().__init__(name)
        self.color = color

    def speak(self):
        return f"{self.name} says Meow!"

# Create instances of the child classes
dog = Dog("Buddy", "Golden Retriever")
cat = Cat("Whiskers", "Tabby")

# Access inherited methods and attributes
print(f"Dog's name: {dog.name}")
print(f"Dog's breed: {dog.breed}")
print(dog.speak())

print(f"\nCat's name: {cat.name}")
print(f"Cat's color: {cat.color}")
print(cat.speak())
````

**Question 10: How can time series analysis be used for anomaly detection?**

**Answer:** Time series analysis is highly effective for anomaly detection by modeling the normal behavior of a sequence of data points collected over time. Anomalies are then identified as deviations from this learned normal pattern. Here's how it can be used:

1.  **Forecasting and Residuals:** A common approach is to build a forecasting model (e.g., ARIMA, Prophet, Exponential Smoothing) on historical time series data. The model predicts future values, and the residuals (the difference between actual and predicted values) are then analyzed. Large residuals or residuals falling outside a predefined confidence interval can indicate an anomaly.
2.  **Statistical Process Control (SPC):** Techniques like control charts (e.g., Shewhart charts, EWMA charts) can be applied to time series data. These charts establish upper and lower control limits based on the historical variability of the data. Any data point that falls outside these limits is considered an anomaly.
3.  **Pattern Recognition:** Time series analysis can identify recurring patterns (seasonality, trend) in the data. Anomaly detection algorithms can then be designed to flag events that break these established patterns. For instance, a sudden, unseasonal spike in sales could be an anomaly.
4.  **Feature Engineering:** Features like rolling means, standard deviations, differences from previous periods, or Fourier transforms can be extracted from time series data. These features can then be fed into traditional anomaly detection algorithms (like Isolation Forest or One-Class SVM) to identify unusual behaviors.
